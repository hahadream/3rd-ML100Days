- 作業目標 : iris (dataset.load_iris()) 資料嘗試跑 kmeans (可以測試不同的群數 , init 等)

```python
from sklearn import datasets

iris = datasets.load_iris()
X = iris.data
y = iris.target
```

### 載入 相關套件 並 執行 kmean 實驗 ...

- 實驗測試不同的群數
- 實驗測試不同的初始值
- 呈現結果

```python
import pandas as pd
import itertools
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
```

```python
from sklearn.cluster import KMeans
%matplotlib inline
```

```python
# Variations in n_clusters
estimators = []
clusters = [2, 3, 5, 8, 10]
for num in clusters: 
    name = str(num) + 'clusters'
    alg = KMeans(n_clusters = num)
    estimators.append((name, alg))
```

```python
fignum = 1 
titles = [estimators[i][0] for i in range(len(estimators))]
for name, est in estimators:
    fig = plt.figure(fignum, figsize = (2,1))
    ax = Axes3D(fig, rect = [0, 0, 8, 8], elev = 48, azim = 134)
    est.fit(X)
    labels = est.labels_
    ax.scatter(X[:,0], X[:,1], X[:,3], c = labels.astype(np.float), edgecolor = 'k')
    ax.w_xaxis.set_ticklabels([])
    ax.w_yaxis.set_ticklabels([])
    ax.w_zaxis.set_ticklabels([])
    ax.set_title(titles[fignum - 1])
    ax.dist = 12
    fignum = fignum + 1 

fig = plt.figure(fignum, figsize = (4,3))
ax = Axes3D(fig, rect = [0,0,8,8], elev = 48, azim = 134)

new_y = np.choose(y, [1,2,0]).astype(np.float)
ax.scatter(X[:,0], X[:,1], X[:,3], c = y, edgecolor = 'k')
ax.w_xaxis.set_ticklabels([])
ax.w_yaxis.set_ticklabels([])
ax.w_zaxis.set_ticklabels([])
ax.set_title('Ground truth')
ax.dist = 12
```

```python
fignum = 1 
for name, est in estimators: 
    fig = plt.figure(fignum, figsize = (4,3))
    est.fit(X)
    labels = est.labels_
    plt.scatter(X[:,0], X[:,1], c = labels.astype(np.float), edgecolor = 'k')
    plt.title(titles[fignum - 1])
    fignum = fignum + 1

fig = plt.figure(fignum, figsize = (4,3))
new_y = np.choose(y, [1,2,0]).astype(np.float)
plt.scatter(X[:,0], X[:,1], c = new_y, edgecolor = 'k')
plt.title('Ground truth')
```
