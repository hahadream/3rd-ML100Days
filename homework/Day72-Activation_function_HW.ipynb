# Rectified Linear Unit- Relu 

f(x)=max(0,x)


```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline

'''
作業:
    寫出 ReLU & dReLU 一階導數
    並列印
'''
```

```python
def relu(x): 
    result = np.zeros(len(x))
    for i in range(len(x)): 
        if x[i] > 0: 
            result[i] = x[i]
    return result

def drelu(x):
    result = np.zeros(len(x))
    for i in range(len(x)):
        if x[i] > 0:
            result[i] = 1
    return result
```

```python
x = np.linspace(-10,10,100)

fig, (ax1, ax2) = plt.subplots(2,1, figsize = (6, 9))
ax1.plot(x, relu(x), 'b')
ax1.set_title("Plot of ReLU")
ax2.plot(x, drelu(x), 'r')
ax2.set_title("Plot of dReLU")
```
